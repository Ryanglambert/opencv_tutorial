```python
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import signal

input_image = cv2.imread('wheel.png', cv2.IMREAD_GRAYSCALE)
```

```python
input_image.shape
```

# 1. Edge detection with sobel operator

```python
y_crop, x_crop = 45, 33
crop_img = input_image[0 + x_crop: 356 - x_crop, 0 + y_crop: 362 - y_crop]

plt.imshow(crop_img, cmap='gray')
```

### Edge detection with sobel operator

```python
ddepth = cv2.CV_64F
dx = 1
dy = 0
sobelx = cv2.Sobel(crop_img, ddepth, dx, dy)

plt.imshow(sobelx, cmap='gray')
```

```python
sobelx = cv2.Sobel(crop_img, cv2.CV_64F, 1, 0)
sobely = cv2.Sobel(crop_img, cv2.CV_64F, 0, 1)

sobel = cv2.sqrt(cv2.addWeighted(cv2.pow(sobelx, 2.0), 1.0, cv2.pow(sobely, 2.0), 1.0, 0.0))
```

```python
plt.figure(figsize=(6, 6));
plt.subplot(2, 2, 1); plt.title('Input image'); plt.axis('off');
plt.imshow(input_image, cmap='gray');
plt.subplot(2, 2, 2); plt.title('Sobel x operation'); plt.axis('off');
plt.imshow(cv2.absdiff(sobelx, 0.0), cmap='gray')
plt.subplot(2, 2, 3); plt.title('Sobel y operation'); plt.axis('off');
plt.imshow(cv2.absdiff(sobely, 0.0), cmap='gray');
plt.subplot(2, 2, 4); plt.title('Sobel intensity'); plt.axis('off');
plt.imshow(sobel, cmap='gray');
```

### Doing the math of the sobel operator on our own
thanks numpy

Math of the Sobel operator from [https://en.wikipedia.org/wiki/Sobel_operator]

```python
G_x = np.array([[1], [2], [1]]).dot(np.array([[1, 0, -1]]))
G_y = np.rot90(G_x, -1)

conv_x = signal.convolve2d(crop_img, G_x, 'valid')
conv_y = signal.convolve2d(crop_img, G_y, 'valid')
```

```python
plt.imshow(conv_x, cmap='gray');
```

```python
plt.imshow(conv_y, cmap='gray');
```

### that's cool but.

The difference matrix highlighted areas based on their rate of change

This is signed change though. The magnitude of that change is more interesting. (gradient magnitude)

That is handled by this next step to enable us to highlight the magnitude of 'edge-ness' themselves.

Gradient magnitude given by: 


```python
sobel_grad_mag = np.sqrt(conv_x**2 + conv_y**2)
plt.imshow(sobel_grad_mag, cmap='gray')
```

### Doing the math of the Canny operator on our own

This is a bit more involved 

make a gaussian filter

```python
std = 1.375
size = 5
gf = signal.windows.gaussian(size, std).reshape(-1, 1)
gf = (gf.dot(gf.T))
print(gf)
```

1. Apply gaussian filter to smooth the image
2. Find intensity gradients like in the sobel filter
3. apply non-maximum suppression to get rid of spurious response to edge detection
4. apply double threshold to determin potential edges
5. track edge by histeresis: Finalize the detection of edges by suppressing all the other edges that are weak and not connected to strong edges.

```python
def gradient(img):
    "simple x, y gradient"
    G_x = np.array([[1], [2], [1]]).dot(np.array([[1, 0, -1]]))
    G_y = np.rot90(G_x, -1)
    conv_x = signal.convolve2d(img, G_x, 'valid')
    conv_y = signal.convolve2d(img, G_y, 'valid')
    return np.sqrt(conv_x**2 + conv_y**2)
```

```python
conv_gauss = signal.convolve2d(sobel_grad_mag, gf, mode='valid')
plt.imshow(conv_gauss, cmap='gray')
```

```python
conv_gauss_grads = gradient(crop_img)

plt.imshow(conv_gauss_grads, cmap='gray')
```
